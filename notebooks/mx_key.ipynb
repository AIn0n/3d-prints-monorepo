```python
from jupyterscad import view
from solid2 import cube
from solid2.extensions.bosl2 import round3d
```

```python
nozzle_size = 0.4
```

```python
mx_stem_l = 4.0
mx_stem_w = 1.2
mx_stem_h = 3.6
mx_stem_l, mx_stem_w, mx_stem_h
```

```python
mx_stem_single = cube(mx_stem_l, mx_stem_w, mx_stem_h, center=True)
view(mx_stem_single)
```

```python
mx_stem = mx_stem_single + mx_stem_single.rotateZ(90)
view(mx_stem)
```

```python
mx_stem_mounting_w = mx_stem_l + nozzle_size * 4
mx_stem_mounting_l = mx_stem_l + nozzle_size * 2
mx_stem_mounting_w, mx_stem_mounting_l
```

```python
mx_stem_mounting = (
    cube([mx_stem_mounting_w, mx_stem_mounting_l, mx_stem_h], center=True) - mx_stem
)
view(mx_stem_mounting)
```

```python
u = 18
keycap_h = 1.2

keycap_base = cube([u, u, keycap_h], center=True)
rounded_keycap = round3d(_or=nozzle_size)(keycap_base)
view(rounded_keycap)
```

```python
final = rounded_keycap + mx_stem_mounting.up(mx_stem_h / 2 + keycap_h / 2)
```

```python
view(final)
```

```python
final.save_as_stl("sg_keycap.stl")
```
