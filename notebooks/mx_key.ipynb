```python
from jupyterscad import view
from solid2 import cube, square
from solid2.extensions.bosl2 import round3d
```

```python
nozzle_size = 0.4
```

```python
mx_stem_l = 4.0
mx_stem_w = 1.2
mx_stem_h = 3.0
mx_stem_l, mx_stem_w, mx_stem_h
```

```python
mx_stem_arm = square([mx_stem_l, mx_stem_w], center=True)
mx_stem_base = mx_stem_arm + mx_stem_arm.rotateZ(90)
```

```python
mx_stem = mx_stem_base.linear_extrude(height=mx_stem_h)
view(mx_stem)
```

```python
mx_stem_mounting_w = mx_stem_l + nozzle_size * 4
mx_stem_mounting_l = mx_stem_l + nozzle_size * 2
mx_stem_mounting_w, mx_stem_mounting_l
```

```python
mx_stem_mounting_base = mx_stem_base.offset(r=nozzle_size)
mx_stem_mounting = mx_stem_mounting_base.linear_extrude(height=mx_stem_h) - mx_stem
view(mx_stem_mounting)
```

```python
from solid2.extensions.bosl2 import round_corners, offset
from solid2 import polygon
```

```python
u = 18
keycap_h = 1
round_r = 2

keycap_base = square([u, u], center=True)
rounded_keycap = polygon(
    round_corners(path=keycap_base, radius=round_r)
).linear_extrude(height=keycap_h)
view(rounded_keycap)
```

```python
final = rounded_keycap + mx_stem_mounting.up(keycap_h - 0.01)
view(final)
```

```python
final.save_as_stl("sg_keycap.stl")
```
