```python
from jupyterscad import view
from solid2 import cube
```

```python
stone_height = 1.5
stone_width = 63.25
stone_len = 152.0
```

```python
added_size = 2
overall_height = 50
```

```python
base = cube(
    [stone_len + added_size, stone_width + added_size, overall_height], center=True
)
```

```python
view(base)
```

```python
offset = 0.2
```

```python
stone_offset = cube(
    [stone_len + offset, stone_width + offset, stone_height], center=True
).up(overall_height / 2)
```

```python
view(stone_offset)
```

```python
view(base - stone_offset)
```

```python
# empty out the inside, to make the print faster

stopper_size = 13


empty_area = cube(
    [
        stone_len + added_size - stopper_size,
        stone_width + added_size - stopper_size,
        overall_height + 10,
    ],
    center=True,
).down(5)
```

```python
view(base - stone_offset - empty_area, outfile="stone-holder.stl")
```
