```python
from jupyterscad import view
from solid2 import cube, cylinder
```

```python
length = 100

base = cube([length,30,10],center=True)
row = cube([length,4, 10], center=True).up(3)
```

```python
rounding_base = cube([length, 10, 10], center=True)
rounding_cyl = cylinder(h=length, r=10, center=True).rotateY(90).translate([0, 5, 5])
rounding_obj = (rounding_base - rounding_cyl).rotateY(180)
view(rounding_obj)
```

```python
view(base - row - rounding_obj.rotate([-90, 0, 0]).translateY(10) - rounding_obj.translateY(-10), outfile='podstawka-fotobutka.stl')
```
